```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import Rectangle
from matplotlib.colors import ListedColormap
```

```python
fig, ax = plt.subplots()
ax.add_patch(Rectangle((0, 1-0.25), 1, 0.5, fill=True, lw=0, color='#2b6a6c'))
ax.add_patch(Rectangle((0, 2-0.25), 2, 0.5, fill=True, lw=0, color='#49b5b8'))
ax.add_patch(Rectangle((1, 1-0.25), 1, 0.5, fill=True, lw=0, color='#163738'))

plt.text(1, 2, "1", horizontalalignment="center", verticalalignment="center", color="w")
plt.text(0.5, 1, "0", horizontalalignment="center", verticalalignment="center", color="w")
plt.text(1.5, 1, "2", horizontalalignment="center", verticalalignment="center", color="w")


plt.xlim(left=0, right=2)
plt.ylim(bottom=0, top=2+1)
plt.yticks([0, 1, 2])
plt.xticks([0, 1, 2])
plt.title('Example Transaction Scheduling Problem Solution')
plt.ylabel('Machines')
plt.xlabel('Time')
plt.savefig('images/base_problem.png')
plt.show()
```

```python
fig, ax = plt.subplots()
arr = np.array([[0, 0, 1], [0, 0, 0], [1, 0, 0]])
cmap = ListedColormap(['#49b5b8', '#163738'])
plt.imshow(arr, cmap=cmap)

cbar = plt.colorbar()
cbar.set_ticks(ticks=[0.25, 0.75], labels=['No Conflict', 'Conflict'])
plt.yticks([0, 1, 2])
plt.xticks([0, 1, 2])
plt.hlines(y=np.arange(0, 2)+0.5, xmin=np.full(2, 0)-0.5, xmax=np.full(2, 3)-0.5, color='white')
plt.vlines(x=np.arange(0, 2)+0.5, ymin=np.full(2, 0)-0.5, ymax=np.full(2, 3)-0.5, color='white')
plt.title('Conflict Matrix')
plt.ylabel('Transaction')
plt.xlabel('Transaction')
plt.tight_layout()
plt.savefig('images/base_problem_conflicts.png')
plt.show()
```
